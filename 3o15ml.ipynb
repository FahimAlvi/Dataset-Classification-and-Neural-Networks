Oversampling and Undersampling:
train dataset labels are mostly zeros, so the data is imbalaced. So oversampling is used to increase samples with '1' label; and undersampling is done to reduce the number of samples with '0' label 

```python
from imblearn.combine import SMOTETomek
smt = SMOTETomek(random_state=42)
x_trainos, y_trainos = smt.fit_resample(x_train, y_train.ravel())
  
print('After OverSampling and Undersampling, the shape of train_X: {}'.format(x_trainos.shape))
print('After OverSampling and Undersampling, the shape of train_y: {} \n'.format(y_trainos.shape))
  
print("After OverSampling and Undersampling, counts of label '1': {}".format(sum(y_trainos == 1)))
print("After OverSampling and Undersampling, counts of label '0': {}".format(sum(y_trainos == 0)))
```

Neural Network Model

```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras import datasets, layers, models, losses, optimizers
import tensorflow as tf
from keras.regularizers import l1


model = Sequential()
model.add(Dense(24, kernel_regularizer=l1(0.02), bias_regularizer=l1(0.02), activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(18, kernel_regularizer=l1(0.02), bias_regularizer=l1(0.02), activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss=tf.keras.losses.BinaryCrossentropy(), optimizer='adam', metrics=[tf.keras.metrics.AUC()])

training_history = model.fit(x_trainos, y_trainos, epochs=100, batch_size=64)

```

```python
#Training accuracy of the model
_, accuracy = model.evaluate(x_trainos, y_trainos)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# Loss over graph

import plotly.graph_objects as go

loss = training_history.history['loss']

fig = go.Figure()
fig.add_trace(go.Scatter(x=np.arange(start=1,stop=len(loss)), y=loss,
                    mode='lines+markers',
                    name='Training Loss'))
fig.update_layout(title = "Training Loss vs Epoch Graph", title_x = 0.5,
                  xaxis_title = 'Epoch',
                  yaxis_title = 'Training Loss',
                  width = 800,
                  height = 500)
fig.show()
```

```python
# Predicting using the model
predicted_classes = model.predict(x_test)
print(predicted_classes.shape, y_test.shape)

predicted_classes = np.argmax(np.round(predicted_classes),axis=1)
predicted_classes.shape
```

```python
#Print the test accuracy
correctly_classified = len(np.where(predicted_classes==y_test)[0])
misclassified = y_test.shape[0] - correctly_classified
test_accuracy = correctly_classified/(correctly_classified + misclassified)*100
print("Test accuracy =",test_accuracy,"%")
print("Correctly CLassified: ",correctly_classified)
print("Misclassified: ",misclassified)
```
